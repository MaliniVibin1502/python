```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
uc= pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data",names= headers)
uc.head(10)
```

Handling missing data in the dataframe
- Convert ? to NaN
- Evaluating Missing data
- Dealing with the missing data

```python
uc.replace("?", np.nan,inplace=True)
uc.head(10)
```

```python
missing_data=uc.isnull()
missing_data.head(10)
```

Counting the missing values for each column

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

Replace by mean:

- "normalized-losses": 41 missing data, replace them with mean
- "stroke": 4 missing data, replace them with mean
- "bore": 4 missing data, replace them with mean
- "horsepower": 2 missing data, replace them with mean
- "peak-rpm": 2 missing data, replace them with mean


Replace by frequency: "num-of-doors": 2 missing data, replace them with "four".


Drop the whole row: "price": 4 missing data, simply delete the whole row


```python
mean_normalized_losses = uc['normalized-losses'].astype(float).mean(axis=0)
print("Mean of Normalized Losses is = ", mean_normalized_losses)
```

```python
uc["normalized-losses"].replace(np.nan, mean_normalized_losses, inplace=True)
uc["normalized-losses"].isnull().value_counts()
```

```python
avg_bore=uc['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
uc['bore'].replace(np.nan,avg_bore,inplace=True)
uc['bore'].isnull().value_counts()
```

```python
avg_stroke=uc['stroke'].astype('float').mean(axis=0)
print("Average of bore:", avg_stroke)
```

```python
uc['stroke'].replace(np.nan,avg_stroke,inplace=True)
uc['stroke'].isnull().value_counts()
```

```python
avg_horsepower=uc['horsepower'].astype('float').mean(axis=0)
print("Average of bore:", avg_horsepower)
```

```python
uc['horsepower'].replace(np.nan,avg_horsepower,inplace=True)
uc['horsepower'].isnull().value_counts()
```

```python
avg_rpm=uc['peak-rpm'].astype('float').mean(axis=0)
print("Average of bore:", avg_rpm)
```

```python
uc['peak-rpm'].replace(np.nan,avg_horsepower,inplace=True)
uc['peak-rpm'].isnull().value_counts()
```

```python
uc['num-of-doors'].value_counts()
```

```python
uc['num-of-doors'].value_counts().idxmax()
```

```python
uc["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
uc.dropna(subset=["price"], axis=0, inplace=True)
uc.reset_index(drop=True, inplace=True)
```

```python
uc.head(10)
```

Check and correct the data types of the data in the dataset

```python
uc.dtypes
```

```python
uc[["bore", "stroke"]] = uc[["bore", "stroke"]].astype("float")
uc[["normalized-losses"]] = uc[["normalized-losses"]].astype("int")
uc[["price"]] = uc[["price"]].astype("float")
uc[["peak-rpm"]] = uc[["peak-rpm"]].astype("float")
```

```python
uc.dtypes
```

Standardising data - Convert mpg to L/100km by mathematical operation (235 divided by mpg)

```python
uc['highway-L/100km'] = 235/uc['highway-mpg']
uc.head()
```

Data Normalization

Scale the columns "length", "width" and "height"  from 0 to 1

```python
uc['length'] = uc['length']/uc['length'].max()
uc['width'] = uc['width']/uc['width'].max()
uc['height'] = uc['height']/uc['height'].max()
```

```python
uc[['length','width','height']]
```

Binning the values into categories for analysis

In the data set, "horsepower" is a real valued variable ranging from 48 to 288 and it has 59 unique values. The further analysis is done for price difference between cars with high horsepower, medium horsepower, and little horsepower (3 types)? Therefore the data horeppwer will be rearranged into three ‘bins' to simplify analysis.



```python
uc["horsepower"]=uc["horsepower"].astype(int, copy=True)
plt.hist(uc["horsepower"])
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("Horsepower bins")
```

Grouping & Analysis

```python
uc['drive-wheels'].unique()
```

```python
uc_group_one = uc[['drive-wheels','body-style','price']]
```

```python
uc_group_one = uc_group_one.groupby(['drive-wheels'],as_index=False).mean()
uc_group_one
```

```python
uc_gptest = uc[['drive-wheels','body-style','price']]
grouped_test1 = uc_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0)
grouped_pivot
```

```python
group_bs=uc[['body-style','price']]
group_uc=group_bs.groupby('body-style',as_index=False).mean()
group_uc
```

Drive Wheels and Body Style vs. Price

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')


row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index


ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)


ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)


plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

Correlation and Causation

- Calculating the Pearson Correlation Coefficient and P-value of 'wheel-base' and 'price'.
- Calculating the Pearson Correlation Coefficient and P-value of 'horsepower' and 'price'.
- Calculating the Pearson Correlation Coefficient and P-value of 'highway-mpg' vs. 'price'

```python
pearson_coef,p_value=stats.pearsonr(uc['wheel-base'],uc['price'])
print("Pearson Coefficient = ",pearson_coef,"with P value of",p_value)
```

```python
pearson_coef,p_value=stats.pearsonr(uc['horsepower'],uc['price'])
print("Pearson Coefficient = ",pearson_coef,"with P value of",p_value)
```

```python
pearson_coef, p_value = stats.pearsonr(uc['highway-mpg'], uc['price'])
print( "The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P = ", p_value )
```
